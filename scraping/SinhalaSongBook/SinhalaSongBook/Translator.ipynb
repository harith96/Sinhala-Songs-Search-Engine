```python
import pandas as pd
import numpy as np
from googletrans import Translator
import re
import os
from math import isnan
```

```python
translator = Translator()
```

```python
df = pd.read_csv("songs.csv")
sample = df.sample(100)
```

```python
def format_names(names):
  parts = str(names).strip().split(",")
  if(len(parts) != 0):
    translated = []
    
    for part in parts:
        translated.append(translator.translate(part,dest="si").text)
    
    if(len(translated)!=1):
       return ",".join(translated)
    else: 
       return translated[0]
```

```python
def format_titles(song):
    if(song["title_en"] != None):
        title_en = str(song["title_en"])
        title_en = title_en.replace('|',',')
        parts = re.split(r',|_|-|!|–', title_en)
        song["title_en"] = parts[0].strip()
        if(song["title_si"] == None or str(song["title_si"]) == 'nan'):
            if(len(parts) == 2):
                song["title_si"] = parts[1].strip()
            else:
                song["title_si"] = translator.translate(parts[0],dest='si').text
                
    return song
```

```python
df["artist_si"] = df["artist"].apply(format_names)
# df["composer_si"] = df["composer"].apply(format_names)
# df["genre_si"] = df["genre"].apply(format_names)
# df["writer"] = df["writer"].apply(format_names)
```

```python
for index, row in df.iterrows():
    df.loc[index] = format_titles(row)
```
