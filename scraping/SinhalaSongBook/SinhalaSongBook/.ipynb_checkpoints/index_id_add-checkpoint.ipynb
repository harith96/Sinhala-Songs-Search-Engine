```python
import pandas as pd
```

```python
df = pd.read_json("lyrics.json")
```

```python
global count
count = 0

def insert_index_id_columns(row):
    global count
    count = 1
    return pd.Series([count, "songs"], index=['song_id', '_index'])

df = pd.concat([df, df.apply(insert_index_id_columns, axis=1)], axis = 1)
```

```python
df.sample(100)
```

```python
df.to_json("lyrics.json", orient='records', index=True)
```
